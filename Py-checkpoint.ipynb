```python
def predict_student_grade(student_data, categorical_cols, label_encoders, features, scaler, best_rf, le_grade):

    try:

        student_df = pd.DataFrame([student_data])
    
        if 'Scholarship' in student_df.columns:
            student_df['Scholarship'] = (
                student_df['Scholarship']
                .astype(str)
                .str.replace('%', '')
                .replace('None', '0')
                .astype(float)
            )
        
        for col in categorical_cols:
            if col in student_df.columns:
                student_df[col] = label_encoders[col].transform(student_df[col])

        student_scaled = scaler.transform(student_df[features])
        
        predicted_grade_encoded = best_rf.predict(student_scaled)
        predicted_grade = le_grade.inverse_transform(predicted_grade_encoded)
        
        return predicted_grade[0]
    
    except Exception as e:
        print(f"Prediction error: {e}")
        return None
```
